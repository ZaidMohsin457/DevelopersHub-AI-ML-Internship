# Task 1: Exploring and Visualizing the Iris Dataset

This notebook demonstrates how to load, inspect, and visualize the Iris dataset using pandas, matplotlib, and seaborn.

## 1. Import Required Libraries
Import pandas, matplotlib, and seaborn for data handling and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better plots
sns.set(style="whitegrid")
```

## 2. Load the Iris Dataset
Load the Iris dataset into a pandas DataFrame using seaborn's `load_dataset` function.

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')

# Display the first few rows
display(iris.head())
```

## 3. Inspect Dataset Structure
Print the shape, column names, and the first few rows of the dataset.

```python
# Print the shape of the dataset
print('Shape:', iris.shape)

# Print the column names
print('Columns:', iris.columns.tolist())

# Display the first 5 rows
iris.head()
```

## 4. Summary Statistics
Use `.info()` and `.describe()` to display summary statistics and data types.

```python
# Display info about the dataset
iris.info()

# Display summary statistics
iris.describe()
```

## 5. Scatter Plot of Feature Relationships
Visualize relationships between features using a pairplot from seaborn.

```python
# Pairplot to visualize relationships between features
sns.pairplot(iris, hue='species', diag_kind='hist', corner=True)
plt.suptitle('Pairplot of Iris Features', y=1.02)
plt.show()
```

## 6. Histograms of Feature Distributions
Plot histograms for each numerical feature to show their value distributions.

```python
# Plot histograms for each numerical feature
iris.hist(bins=20, figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Iris Features', y=1.02)
plt.tight_layout()
plt.show()
```

## 7. Box Plots for Outlier Detection
Create box plots for each feature to identify potential outliers.

```python
# Create box plots for each numerical feature
features = iris.columns[:-1]  # Exclude species column
plt.figure(figsize=(12, 8))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=iris[feature], color='lightgreen')
    plt.title(f'Box Plot of {feature}')
plt.tight_layout()
plt.show()
```
